```python
import pandas as pd
```

```python
data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data') 
data.head()
```

```python
data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',chunksize=10 ) # read in chunks of 10

for chunk in data:
  print(chunk)
```
